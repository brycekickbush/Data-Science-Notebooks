```python
#q2
ipList = logFile['ip'].tolist() #Get all IP's onto a list
ipUsageDict = dict()
for i in ipList:
    ipUsageDict[i] = ipUsageDict.get(i, 0) + 1 #Create a dictionary with a key being IP, and value being a counter

sortedValues = sorted(ipUsageDict.values(), reverse=True) #Sort dictionary in descending order USE FOR Q6
topList = dict()
for i in range(10): #Top 10 values
    timesAccessed = sortedValues[i]
    for x in ipUsageDict: #If ip is one of the top 10 add to a new dict
        if ipUsageDict[x] == timesAccessed:
            topList[x] = timesAccessed     
topList
```

```python
#q4

extentionList = logFile['extention'].tolist()
cikList = logFile['cik'].tolist()
accessionList = logFile['accession'].tolist()
pathDict = dict()
pathList = []
for i in range(len(extentionList)):
    size = len(str(cikList[i]))
    path = str(cikList[i])[:size-2] + "/" + accessionList[i] + "/" + extentionList[i]
    pathList.append(path)
    pathDict[path] = pathDict.get(path, 0) + 1
sortedList = sorted(pathDict.values(), reverse=True)

topExtention = ""
for x in pathDict:
    if pathDict[x] == sortedList[0]:
        topExtention = x
final = topExtention
final
```

```python
#q7
#Fraction of high volume users
#Check ip location, dict will contain ip and number of times accessed.
#Dictionary ip:counter accessed
#Loop through ip's, country:(ip:times accessed) -- > nested dictionary
#Country -----> dict of ip:times Accessed specific countrty
#loop through the ip's --> check country ---> build dictonary with key being country
#Loop through keys within each nested dictionary. 
countryNestedDict = dict()
for i in ipUsageDict:
    if edgar_utils.lookup_region(i) not in countryNestedDict:
        countryNestedDict[edgar_utils.lookup_region(i)] = {i:ipUsageDict[i]}
    countryNestedDict[edgar_utils.lookup_region(i)].update({i:ipUsageDict[i]})
highVolDict = dict() #Finald dictionary

for x in countryNestedDict: #Each key in dictionary
    ipDict = countryNestedDict[x]
    numIps = len(ipDict.keys())
    numHighVol = 0
    numAccessed = ipDict.values()
    for y in numAccessed:
        if y >= 1000:
            numHighVol += 1
    frac = numHighVol/numIps
    highVolDict[x] = frac
highVolDict
```
